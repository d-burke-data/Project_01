```python
import pandas as pd
```

```python
input_file_path = "Data/home_value.csv"
output_file_path = "Data/home_value_trim_and_format.csv"

original_df = pd.read_csv(input_file_path)
original_df.head()
print(original_df.dtypes)
```

```python
# Split date into year and month. Day is irrelevant.

clean_df = original_df.copy()
clean_df.insert(1, 'Year', "")
clean_df.insert(2, 'Month', "")
clean_df['Date'] = clean_df['Date'].str[0:7]
clean_df['Year'] = clean_df['Date'].str[0:4]
clean_df['Month'] = clean_df['Date'].str[5:7]
clean_df.head()
```

```python
clean_df.columns
```

```python
# Output file

clean_df.to_csv(output_file_path, index=False)
```
